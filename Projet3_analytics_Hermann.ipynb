#### - utilisons un test statistique pour tester s'il ya égalité des moyennes des ventes chez les hommes et chez les femmes  
<b> Hyphothèse H0 <<Les moyennes sont égales>> </b>  
<b> Hyphothèse H1 <<Les moyennes ne sont pas égales>> </b>  
Ici on veut comparer les moyennes de deux variables, donc il faut utiliser le test de student avec la méthode ttest_ind. Pour ce faire il faut vérifier si ces deux variables suivent la loi normale(test de shapiro) et s'ils ont la meme variance(test de levene)

Si la p_value est inferieur a 0.05 on rejette l'hypothèse H0 et on accepte l'hypothèse H1. Dans le cas contraire on accepte l'hypothèse H0.  

```python
sous_data_sex1=data[data['sex']=='f']
sous_data_sex2=data[data['sex']=='m']
```

premièrement : On va vérifier si les deux variavles suivent la loi normale

```python
stats.shapiro(sous_data_sex1['price'])
```

```python
stats.shapiro(sous_data_sex2['price'])
```

```python
stats.levene(sous_data_sex2['price'],sous_data_sex1['price'],center='median')
```

<b>conclusion</b>: Donc ces deux variables  suivent  la loi normale mais pas la meme variance. On va donc utiliser le test  non paramétrique de  mannwhitneyu.

```python
#y = stats.ttest_ind(sous_data_sex2['price'],sous_data_sex1['price'])
#print(y)
```

```python
#stats.wilcoxon(sous_data_sex1['price'],sous_data_sex2['price'])
```

```python
stats.mannwhitneyu(sous_data_sex1['price'],sous_data_sex2['price'])
```

<b>conclusion</b>: La pvalue < 0.05 donc l'hypothèse H0 est rejeté par conséquent il  ya égalité entre les deux moyennes

#### Indice de GUINI

```python
def gini (donne) :
    i=range(1,donne.shape[0]+1,1)
    n=donne.shape[0]+1
    donne=np.sort(donne)
    prod=i*donne
    ginivalue=(2*prod.sum())/(n*donne.sum())-(n+1)/n
    return ginivalue
print(gini(venteparclient['price']))
```

<b>Conclusion : </b> La courbe de Lorenz est éloignée de la droite d'équirépartition et l'indice ou statistique de GINI=0,44 qui est très éloigné de 1 Donc  L'inégalité entre le chiffre d'affaire par client est  importante.

On se pose la question de savoir s'il ya une relation entre le genre d'un client et la catégorie de produit acheté . Pour répondre à cette question, nous allons utiliser les tests statistiques.  
<b>Hypothèse H0 :</br> << ces deux variables sont indépendantes >>   
<b>Hypothèse H1 :</br> << ces deux variables sont dépendantes >>  
Comme il s'agit d'un test statistique entre deux variables catégorielles nous allons utiliser le test de CHI2

```python
from scipy.stats import chi2_contingency
import scipy.stats as stats 
```

#### Observons la repartition de ces deux variables.

```python
sns.countplot(data=data,x='sex',hue='categ')
plt.title('Observation graphique de la Relation entre le sexe et la catégorie du livre acheté')
```

#### TEST de CHI2

```python
tablecc=pd.crosstab(data['sex'],data['categ'])
print(tablecc)
result=chi2_contingency(tablecc,correction=True)
print('La statitique de test :',result[0])
print('La p_valeur est :',result[1])
print('Le degré de liberté est :',result[2])
```

#### Observons les différentes corrélation de la crosstable

```python
sns.heatmap(tablecc,annot=True,cmap='YlGnBu')
```

<b>Conclusion : </b> La p_valeur est inférieur à 0,05  donc l'hypothèse H0 est rejetée :  ces deux variables sont dépendantes(liées) d'ou il ya un lien entre le sex du client et la catégorie du livre acheté. Donc La catégorie du livre achetée dépend du sexe du client (Le sex a une influence sur la catégorie du livre acheté). 

#### Corrélation entre les deux variables en utilisant le V de cramer

```python
n = sum(tablecc.sum())
dim_min1 = min(tablecc.shape)-1
x2=result[0]
V_cramer = np.sqrt((x2/n)/dim_min1)

print("Le V de Cramer est de :", V_cramer)

#data['sex'].replace(['f','m'],['0','1'],inplace=True)
#data['sex']=pd.to_numeric(data['sex'])
#data['sex'].unique()
#print('Le coefficient de correlation entre ces deux variables est de :',np.corrcoef(data['categ'],data['sex']))
```
