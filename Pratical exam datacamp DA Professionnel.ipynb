#### Indice de Gini

```python
def gini (donne) :
    i=range(1,donne.shape[0]+1,1)
    n=donne.shape[0]+1
    donne=np.sort(donne)
    prod=i*donne
    ginivalue=(2*prod.sum())/(n*donne.sum())-(n+1)/n
    return ginivalue
print(gini(df['revenue']))
```

<b>conclusion</b>: The Lorenz curve is far from the equidistribution line and the GINI index or statistic = 0.26 which is very far from 1. Therefore, the inequality between the revenue per customer is important.

#### Question 4 : other differences between the customers in each group

<b>Hypothse H0 :</br> << ces trois variables sont indpendantes >>   
<b>Hypothse H1 :</br> << ces trois variables sont dpendantes >> 

```python
import scipy.stats as stats 
```

```python
df_E=df.loc[df['sales_method']=='Email',['revenue']]
df_E_C=df.loc[df['sales_method']=='Email + Call',['revenue']]
df_C=df.loc[df['sales_method']=='Call',['revenue']]
```

```python
df_E.shape
```

```python
df_E_C.shape
```

```python
df_C.shape
```

```python
stats.shapiro(df_E['revenue'])
```

```python
stats.shapiro(df_E_C['revenue'])
```

```python
stats.shapiro(df_C['revenue'])
```

<b>Conclusion : </b> The pvalue<0.05 means that these two variables do not follow the normal distribution, so we will use the non-parametric Kruskal wallis test

```python
from scipy.stats import kruskal
vect=kruskal(df_E['revenue'],df_C['revenue'],df_E_C['revenue'])
print(vect)
```

<b>Conclusion : </b> The p-value is less than 0.05, so the H0 hypothesis is rejected and there is a difference between the different groups in terms of income.
